```python
import os
import json
import random
import warnings
warnings.filterwarnings("ignore")

# Deep learning library used for model training and inference.
import tensorflow as tf
# Print available GPUs to confirm hardware setup.
print("GPU:", tf.config.list_physical_devices('GPU'))

import numpy as np
from tensorflow.keras import layers, callbacks
from tensorflow.keras.applications import EfficientNetB0
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# ============================================================
# SHARED CONSTANTS
# ============================================================
SR          = 22050
N_MELS      = 128
N_FFT       = 2048
HOP_LENGTH  = 512
CHUNK_SEC   = 4.0
IMG_SIZE    = (224, 224)
MEL_DB_MIN  = -80.0
MEL_DB_MAX  =   0.0
TOP_DB      = 30

# ============================================================
# TRAINING CONFIG
# ============================================================
# Input dataset and output directory paths.
DATA_DIR      = "/kaggle/input/datasets/salonikarapurkar/goa-bird-spectrograms"
MODEL_DIR     = "/kaggle/working/model_output"
MAX_PER_CLASS = 3500

# Core training hyperparameters.
BATCH_SIZE   = 64
EPOCHS       = 60
LR_HEAD      = 3e-4
LR_FINETUNE  = 5e-5
LR_FULL      = 1e-5
VAL_SPLIT    = 0.15
TEST_SPLIT   = 0.10
SEED         = 42
L2_REG       = 1e-4

# Fixed class list used for labels and reporting.
SPECIES = [
    "Asian_Koel",
    "Indian_Peafowl",
    "Coppersmith_Barbet",
    "Jungle_Babbler",
    "Purple-rumped_Sunbird",
    "Greater_Racket-tailed_Drongo",
    "Indian_Pitta",
    "White-throated_Kingfisher",
    "Red-wattled_Lapwing",
    "Malabar_Whistling_Thrush",
]
NUM_CLASSES = len(SPECIES)

# Create output folder and set random seeds for reproducibility.
os.makedirs(MODEL_DIR, exist_ok=True)
tf.random.set_seed(SEED)
np.random.seed(SEED)

# ============================================================
# STEP 1 â€” LOAD IMAGE PATHS
# ============================================================
print("[LOAD] Scanning spectrogram folders...")

# Collect image file paths and numeric labels.
all_paths, all_labels = [], []

for label_idx, species in enumerate(SPECIES):
    folder = Path(DATA_DIR) / species
    if not folder.exists():
        print(f"  âš ï¸  Missing folder: {folder}")
        continue
    images = list(folder.glob("*.png"))
    random.seed(SEED)
    random.shuffle(images)
    images = images[:MAX_PER_CLASS]
    all_paths.extend([str(p) for p in images])
    all_labels.extend([label_idx] * len(images))
    print(f"  {species.replace('_',' '):<35}  {len(images):>5} images")

all_paths, all_labels = shuffle(all_paths, all_labels, random_state=SEED)
all_labels_arr = np.array(all_labels)

print(f"\n  Total: {len(all_paths)} images  |  {NUM_CLASSES} classes")

label_map = {i: s.replace("_", " ") for i, s in enumerate(SPECIES)}
with open(os.path.join(MODEL_DIR, "label_map.json"), "w") as f:
    json.dump(label_map, f, indent=2)
print(f"  [OK] label_map.json saved")

# ============================================================
# STEP 2 â€” STRATIFIED SPLITS
# ============================================================
# First split: hold out final test set.
X_temp, X_test, y_temp, y_test = train_test_split(
    all_paths, all_labels_arr,
    test_size=TEST_SPLIT,
    stratify=all_labels_arr,
    random_state=SEED
)
# Second split: create validation set from remaining data.
adjusted_val = VAL_SPLIT / (1.0 - TEST_SPLIT)
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp,
    test_size=adjusted_val,
    stratify=y_temp,
    random_state=SEED
)

print(f"\n  Train : {len(X_train)}")
print(f"  Val   : {len(X_val)}")
print(f"  Test  : {len(X_test)}")

# ============================================================
# STEP 3 â€” CLASS WEIGHTS
# ============================================================
# Compute class weights so minority classes get higher loss weight.
class_weights_arr = compute_class_weight(
    class_weight="balanced",
    classes=np.arange(NUM_CLASSES),
    y=y_train
)
class_weight_dict = {i: float(w) for i, w in enumerate(class_weights_arr)}
print(f"\n  Class weights: { {k: round(v,3) for k,v in class_weight_dict.items()} }")

# ============================================================
# STEP 4 â€” tf.data PIPELINE  (clean, no MixUp)
# ============================================================
# Let TensorFlow tune pipeline parallelism automatically.
AUTO = tf.data.AUTOTUNE


# Read, decode, resize, and cast PNG spectrogram images.
def load_and_decode(path, label):
    raw = tf.io.read_file(path)
    img = tf.image.decode_png(raw, channels=3)
    img = tf.image.resize(img, IMG_SIZE)
    img = tf.cast(img, tf.float32)   # NO /255.0 â€” EfficientNet does this internally
    return img, label


# Lightweight image-space augmentations for regularization.
def augment_tf(img, label):
    img = tf.image.random_brightness(img, max_delta=20.0)  # scaled for 0-255
    img = tf.image.random_contrast(img, lower=0.92, upper=1.08)
    crop_w = tf.random.uniform(
        [], minval=IMG_SIZE[1] - 16, maxval=IMG_SIZE[1], dtype=tf.int32
    )
    img = tf.image.random_crop(img, size=[IMG_SIZE[0], crop_w, 3])
    img = tf.image.resize(img, IMG_SIZE)
    img = tf.image.random_flip_left_right(img)
    img = tf.clip_by_value(img, 0.0, 255.0)  # scaled for 0-255
    return img, label


# Build tf.data pipeline with optional augmentation and shuffling.
def make_dataset(paths, labels, training: bool = False) -> tf.data.Dataset:
    ds = tf.data.Dataset.from_tensor_slices((paths, labels))
    ds = ds.map(load_and_decode, num_parallel_calls=AUTO)
    if training:
        ds = ds.map(augment_tf, num_parallel_calls=AUTO)
        ds = ds.shuffle(buffer_size=2000, seed=SEED)
    ds = ds.batch(BATCH_SIZE).prefetch(AUTO)
    return ds


# Build train/validation datasets.
train_ds = make_dataset(X_train, y_train, training=True)
val_ds   = make_dataset(X_val,   y_val,   training=False)


def load_test(path, label):
    raw = tf.io.read_file(path)
    img = tf.image.decode_png(raw, channels=3)
    img = tf.image.resize(img, IMG_SIZE)
    img = tf.cast(img, tf.float32)   # NO /255.0
    return img, label

# Build test dataset without augmentation.
test_ds = (
    tf.data.Dataset
    .from_tensor_slices((X_test, y_test))
    .map(load_test, num_parallel_calls=AUTO)
    .batch(BATCH_SIZE).prefetch(AUTO)
)

# ============================================================
# STEP 5 â€” MODEL
# ============================================================
print("\n[BUILD] Building model...")

regularizer = tf.keras.regularizers.l2(L2_REG)


# Build transfer-learning model on top of EfficientNetB0.
def build_model() -> tf.keras.Model:
    base = EfficientNetB0(
        include_top=False,
        weights="imagenet",
        input_shape=(*IMG_SIZE, 3),
        pooling=None
    )
    base.trainable = False

    inputs = tf.keras.Input(shape=(*IMG_SIZE, 3))
    feat   = base(inputs, training=False)

    avg_p  = layers.GlobalAveragePooling2D()(feat)
    max_p  = layers.GlobalMaxPooling2D()(feat)
    x      = layers.Concatenate()([avg_p, max_p])

    x = layers.BatchNormalization()(x)
    x = layers.Dense(512, kernel_regularizer=regularizer)(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("swish")(x)
    x = layers.Dropout(0.50)(x)
    x = layers.Dense(256, kernel_regularizer=regularizer)(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation("swish")(x)
    x = layers.Dropout(0.35)(x)
    outputs = layers.Dense(NUM_CLASSES, activation="softmax")(x)

    return tf.keras.Model(inputs, outputs)


model = build_model()
model.summary()

# SparseCategoricalCrossentropy â€” works with integer labels directly
# Integer labels are used directly with sparse categorical loss.
loss_fn = tf.keras.losses.SparseCategoricalCrossentropy()

# ============================================================
# CALLBACKS
# ============================================================

# Shared callbacks for early stop, LR scheduling, and best checkpoint.
def make_callbacks(tag: str) -> list:
    return [
        callbacks.EarlyStopping(
            monitor="val_accuracy",
            patience=8,
            restore_best_weights=True,
            verbose=1
        ),
        callbacks.ReduceLROnPlateau(
            monitor="val_loss",
            factor=0.5,
            patience=3,
            min_lr=1e-8,
            verbose=1
        ),
        callbacks.ModelCheckpoint(
            filepath=os.path.join(MODEL_DIR, f"best_{tag}.keras"),
            monitor="val_accuracy",
            save_best_only=True,
            verbose=1
        ),
    ]

# ============================================================
# PHASE 1 â€” Head only (10 epochs, base frozen)
# ============================================================
print("\n[PHASE 1]  Head-only training (base frozen)...")

model.compile(
    optimizer=tf.keras.optimizers.Adam(LR_HEAD, clipnorm=1.0),
    loss=loss_fn,
    metrics=["accuracy"]
)

# Phase 1: train only classifier head.
hist1 = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=10,
    class_weight=class_weight_dict,
    verbose=1
)

# ============================================================
# PHASE 2 â€” Fine-tune top 40 layers
# ============================================================
print("\n[PHASE 2]  Fine-tuning top 40 layers...")

# Access backbone model inside the full network.
base_model = model.layers[1]
# Enable backbone, then limit training to top layers for this phase.
base_model.trainable = True
# Phase 2: unfreeze only top layers for controlled fine-tuning.
for layer in base_model.layers[:-40]:
    layer.trainable = False
for layer in base_model.layers:
    if isinstance(layer, layers.BatchNormalization):
        layer.trainable = False

model.compile(
    optimizer=tf.keras.optimizers.Adam(LR_FINETUNE, clipnorm=1.0),
    loss=loss_fn,
    metrics=["accuracy"]
)

# Phase 2: fine-tune top backbone layers.
hist2 = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS,
    callbacks=make_callbacks("phase2"),
    class_weight=class_weight_dict,
    verbose=1
)

# ============================================================
# PHASE 3 â€” Full model fine-tune
# ============================================================
print("\n[PHASE 3]  Full model fine-tune...")

# Phase 3: full-model fine-tuning at a lower learning rate.
base_model.trainable = True
for layer in base_model.layers:
    if isinstance(layer, layers.BatchNormalization):
        layer.trainable = False

model.compile(
    optimizer=tf.keras.optimizers.Adam(LR_FULL, clipnorm=1.0),
    loss=loss_fn,
    metrics=["accuracy"]
)

# Phase 3: final fine-tuning pass.
hist3 = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS,
    callbacks=make_callbacks("final"),
    class_weight=class_weight_dict,
    verbose=1
)

# ============================================================
# STEP 6 â€” EVALUATE
# ============================================================
print("\n[EVAL]  Test set evaluation...")
test_loss, test_acc = model.evaluate(test_ds, verbose=0)
print(f"  Test accuracy : {test_acc*100:.2f}%")
print(f"  Test loss     : {test_loss:.4f}")

# Predict on test set for detailed metrics and plots.
y_pred_probs = model.predict(test_ds, verbose=0)
y_pred       = np.argmax(y_pred_probs, axis=1)
y_true       = y_test
species_names = [s.replace("_", " ") for s in SPECIES]

print("\n  Classification Report:")
print(classification_report(y_true, y_pred, target_names=species_names, digits=3))

# Build confusion matrix to inspect per-class errors.
cm = confusion_matrix(y_true, y_pred)
fig, ax = plt.subplots(figsize=(12, 10))
sns.heatmap(
    cm, annot=True, fmt="d", cmap="Greens",
    xticklabels=species_names, yticklabels=species_names,
    ax=ax, linewidths=0.4
)
ax.set_xlabel("Predicted", fontsize=11)
ax.set_ylabel("Actual",    fontsize=11)
ax.set_title("Confusion Matrix â€” Test Set", fontsize=13)
plt.xticks(rotation=40, ha="right", fontsize=8)
plt.yticks(rotation=0,  fontsize=8)
plt.tight_layout()
cm_path = os.path.join(MODEL_DIR, "confusion_matrix.png")
plt.savefig(cm_path, dpi=150)
plt.close()
print(f"  [OK] Confusion matrix â†’ {cm_path}")

# ============================================================
# STEP 7 â€” TRAINING CURVES
# ============================================================

# Merge phase histories so we can draw one combined curve.
def concat_history(*hists) -> dict:
    combined = {}
    for h in hists:
        for key, vals in h.history.items():
            combined.setdefault(key, []).extend(vals)
    return combined


full = concat_history(hist1, hist2, hist3)
# Mark where each training phase ends on the curve plot.
phase_boundaries = [
    len(hist1.history["loss"]),
    len(hist1.history["loss"]) + len(hist2.history["loss"]),
]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
epochs_range = range(1, len(full["accuracy"]) + 1)

ax1.plot(epochs_range, full["accuracy"],     label="Train Acc", linewidth=2)
ax1.plot(epochs_range, full["val_accuracy"], label="Val Acc",   linewidth=2)
for pb in phase_boundaries:
    ax1.axvline(x=pb, color="gray", linestyle="--", linewidth=1)
ax1.set_title("Accuracy")
ax1.set_xlabel("Epoch")
ax1.legend()
ax1.grid(True, alpha=0.4)

ax2.plot(epochs_range, full["loss"],     label="Train Loss", linewidth=2)
ax2.plot(epochs_range, full["val_loss"], label="Val Loss",   linewidth=2)
for pb in phase_boundaries:
    ax2.axvline(x=pb, color="gray", linestyle="--", linewidth=1,
                label="Phase boundary" if pb == phase_boundaries[0] else "")
ax2.set_title("Loss")
ax2.set_xlabel("Epoch")
ax2.legend()
ax2.grid(True, alpha=0.4)

plt.suptitle("Bird Audio Classifier â€” Training History", fontsize=14)
plt.tight_layout()
curve_path = os.path.join(MODEL_DIR, "training_curves.png")
plt.savefig(curve_path, dpi=150)
plt.close()
print(f"  [OK] Training curves â†’ {curve_path}")

# ============================================================
# STEP 8 â€” SAVE MODEL + CONFIG
# ============================================================
final_path = os.path.join(MODEL_DIR, "bird_classifier_final.keras")
model.save(final_path)
print(f"  [OK] Final model â†’ {final_path}")

# Save model metadata needed by the Streamlit app.
config = {
    "img_size"     : list(IMG_SIZE),
    "species"      : SPECIES,
    "label_map"    : label_map,
    "num_classes"  : NUM_CLASSES,
    "sr"           : SR,
    "n_mels"       : N_MELS,
    "n_fft"        : N_FFT,
    "hop_length"   : HOP_LENGTH,
    "chunk_sec"    : CHUNK_SEC,
    "top_db"       : TOP_DB,
    "mel_db_min"   : MEL_DB_MIN,
    "mel_db_max"   : MEL_DB_MAX,
    "test_accuracy": round(float(test_acc), 4),
}
config_path = os.path.join(MODEL_DIR, "model_config.json")
with open(config_path, "w") as f:
    json.dump(config, f, indent=2)
print(f"  [OK] Config â†’ {config_path}")

print("\n[DONE] Files saved to /kaggle/working/model_output/:")
print("   â€¢ best_phase2.keras")
print("   â€¢ best_final.keras")
print("   â€¢ bird_classifier_final.keras")
print("   â€¢ label_map.json")
print("   â€¢ model_config.json")
print("   â€¢ training_curves.png")
print("   â€¢ confusion_matrix.png")
```
